```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('insurance.csv')
df
```

```python
df.isnull().sum()
```

```python
df1 = df.copy()
df2 = df.copy()
df3 = df.copy()
df4 = df.copy()
```

# Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le
```

```python
cols = df1.columns
cols
```

```python
df1.head()
```

```python
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
for col in cols:
    if is_numeric_dtype(df1[col]):
        continue
    else:
        df1[col] = le.fit_transform(df1[col])
```

```python
df1
```
